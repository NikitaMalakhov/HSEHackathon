```python
df = pd.read_csv('./train.csv')
```

```python
category_replace_map = {
    '?': 0,
    'Communication': 1,
    'Quality': 2,
    'Price': 3,
    'Safety': 4
}

sentiment_replace_map = {
    '+': 2,
    '?': 1,
    '−': 0,
}
```

```python
df["1category"] = df["1category"].map(category_replace_map)
```

```python
df
```

```python
df["sentiment"] = df["sentiment"].map(sentiment_replace_map)
```
